```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh
```

```python
L = 100/0.0529177  #a.u.
HBAR = 1 # a.u.
MASS = 0.067  #mass [a.u]
OMEGA = 10/27211.6  #a.u.
N = 20
```

```python
j = np.arange(0, N, 1)
HAMILTONIAN = np.zeros((N, N))
Ψ = np.zeros(N)
x = np.linspace(0, L, N)
```

```python
def get_energy(j:int) -> float:
    return (HBAR**2 * np.pi**2 * j**2)/(2 * MASS * L**2)

def get_potential(i:int, j:int) -> float:
    i=i+1
    j=j+1
    if i == j:
        return 1/24 * (L**2 *(i**2 * np.pi**2 - 6)*MASS*OMEGA**2)/(i**2 * np.pi**2)
    else:
        return 2 * (L**2 * i * j * MASS * OMEGA**2 * ((-1)**(i+j)+1))/(np.pi**2 *(-i+j)**2 * (i+j)**2)

def kronecker_delta(i:int, j:int) -> int:
    if i==j:
        return 1
    else:
        return 0

def get_φ(l:float, j:int, x_idx:float) -> float:
    return np.sqrt(2/l)*np.sin((j*np.pi*x_idx)/l)
```

```python
for j in range(0, N):
    for i in range(0, N):
        # HAMILTONIAN[i,j] =  np.nan
        HAMILTONIAN[j,i] = get_energy(i)*kronecker_delta(j,i)+get_potential(i,j)

# HAMILTONIAN
```

```python
e, c = eigh(HAMILTONIAN)
```
